```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import context
from numlabs.lab5.lab5_funs import Integrator
from collections import namedtuple
import timeit
# using this to time our code accurately
%alias_magic t timeit 
```

```python
class Integ56(Integrator):
    def set_yinit(self):
        #
        # read in 'albedo_white chi S0 L albedo_black R albedo_ground'
        #
        uservars = namedtuple('uservars', self.config['uservars'].keys())
        self.uservars = uservars(**self.config['uservars'])
        #
        # read in 'whiteconc blackconc'
        #
        initvars = namedtuple('initvars', self.config['initvars'].keys())
        self.initvars = initvars(**self.config['initvars'])
        self.yinit = np.array(
            [self.initvars.whiteconc, self.initvars.blackconc,self.initvars.rabbitpop,self.initvars.foxpop])
        self.nvars = len(self.yinit)
        return None

    def __init__(self, coeff_file_name):
        super().__init__(coeff_file_name)
        self.set_yinit()

    def find_temp(self, yvals):
        """
            Calculate the temperatures over the white and black daisies
            and the planetary equilibrium temperature given the daisy fractions
            
            input:  yvals -- array of dimension [2] with the white [0] and black [1]
                    daisy fractiion
            output:  white temperature (K), black temperature (K), equilibrium temperature (K)
        """
        sigma = 5.67e-8  # Stefan Boltzman constant W/m^2/K^4
        user = self.uservars
        bare = 1.0 - yvals[0] - yvals[1]
        albedo_p = bare * user.albedo_ground + \
            yvals[0] * user.albedo_white + yvals[1] * user.albedo_black
        Te_4 = user.S0 / 4.0 * user.L * (1.0 - albedo_p) / sigma
        temp_e = Te_4**0.25
        eta = user.R * user.L * user.S0 / (4.0 * sigma)
        temp_b = (eta * (albedo_p - user.albedo_black) + Te_4)**0.25
        temp_w = (eta * (albedo_p - user.albedo_white) + Te_4)**0.25
        return (temp_w, temp_b, temp_e)

    def derivs5(self, y, t):
        """y[0]=fraction black daisies
           y[2]=fraction white daisies
           y[1]=rabbit population
           y[3]=fox population
           no feedback between daisies and
           albedo_p (set to ground albedo)
        """
        temp_w, temp_b, temp_e = self.find_temp(y)

        if (temp_b >= 277.5 and temp_b <= 312.5):
            beta_b = 1.0 - 0.003265 * (295.0 - temp_b)**2.0
        else:
            beta_b = 0.0

        if (temp_w >= 277.5 and temp_w <= 312.5):
            beta_w = 1.0 - 0.003265 * (295.0 - temp_w)**2.0
        else:
            beta_w = 0.0
        user = self.uservars
        bare = 1.0 - y[0] - y[1]
        chi = user.tau * y[2]
        gamma = user.phi * (y[1]+y[0]) - user.delta * y[3]
        psi = user.omega*y[2] - user.eta * y[3]
        # create a 1 x 2 element vector to hold the derivitive
        f = np.empty_like(y)
        f[0] = y[0] * (beta_w * bare - chi)
        f[1] = y[1] * (beta_b * bare - chi)
        f[2] = gamma
        f[3] = psi
        return f
```

$\omega$ is the constant of proportionality between rabit population and fox birth rate. $\delta$ is the constant of proportionality between fox population and death rate of rabbits. $\eta$ is the death rate of foxes.

```python
theSolver = Integ56('foxes.yaml')
timeVals,yVals,errorlist = theSolver.timeloop5Err()
yvals = pd.DataFrame.from_records(yVals, columns = ['white', 'black', 'rabbit', 'fox'])
```

```python
thefig, theAx = plt.subplots(1, 1)

points, = theAx.plot(timeVals, yvals['white'], '-b+', label='white daisies')
points.set_markersize(12)
theLine1, = theAx.plot(timeVals, yvals['black'], '--ko', label='black daisies')
theAx.set_title('Predator')
theAx.set_xlabel('time')
theAx.set_ylabel('fractional coverage')
out = theAx.legend(loc='best')
```

```python
plt.plot(timeVals,yvals['rabbit'], '-gx')
plt.title('rabbit population')
plt.ylabel('Rabbits')
plt.xlabel('Time')
plt.axhline(0,c = 'r')
```

```python
plt.plot(timeVals,yvals['fox'], '-yx')
plt.title('Fox population')
plt.ylabel('Foxes')
plt.xlabel('Time')
plt.axhline(0,c = 'r')
```

These is are the most interesting nonzero plots for fox and rabbit populations. We can see that they both oscillate around a steady state before reaching their steady state. What is even more interesting is the foxes in this model have a higher steady state population than the rabbits. Usually in nature, the predators will have a smaller population size than the prey, so clearly this model is not accurate to real life. Another interesting part of this is the daisies appear to reach a more steady state before the rabbit and fox populations, with much smaller oscillations around the steadt state after about 20 years.  
Most other paramter spaces sent either the rabbit or fox populations to a negative number, ie population die out, which means that all animals of the species die so their steady state would then be zero.
